```python
import pickle
from utils import *

# Replace 'your_file.pk' with the actual file path
file_path = 'dataset/meas_14_7/data/DATA_MACHINE_7-14/MEAS_1'
files = get_all_file_paths(file_path)
```

```python
selected = random.sample(files, 4)
selected_img = []

for i in selected:
    with open(i, 'rb') as file:
        data = pickle.load(file)
        selected_img.append(data['img'])

# Create a figure and axes with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2)

# Plot each image on a separate subplot
axs[0, 0].imshow(selected_img[0], interpolation='none', cmap='viridis')
axs[0, 0].set_title('Image 1')
axs[0, 1].imshow(selected_img[1], interpolation='none', cmap='gray')
axs[0, 1].set_title('Image 2')
axs[1, 0].imshow(selected_img[2], interpolation='none', cmap='gray')
axs[1, 0].set_title('Image 3')
axs[1, 1].imshow(selected_img[3], interpolation='none', cmap='gray')
axs[1, 1].set_title('Image 4')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

# check channels

```python
'''Using openCV to read the image'''

import cv2

path = 'dataset/meas_7_7/procIMGs_2/'
files = get_all_file_paths(path)

image = cv2.imread(files[100], cv2.IMREAD_UNCHANGED)  # add cv2.IMREAD_UNCHANGED flag
print('The dimension of the image is:', image.shape)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Plot the image
plt.imshow(image)
plt.axis('off')  # Turn off axis
plt.show()

# image = cv2.imread(files[100])
# print('The dimension of the image is:', image.shape)
```

```python
'''Check its RBG channels'''

# Extract channels
# OpenCV loads an image in BGR format, so:
# Blue channel
blue_channel = image[:, :, 0]
# Green channel
green_channel = image[:, :, 1]
# Red channel
red_channel = image[:, :, 2]

# Create a figure and axes with 2 rows and 2 columns
fig, axs = plt.subplots(3, 1, figsize=(15, 8))

# Plot each image on a separate subplot
axs[0].imshow(blue_channel, interpolation='none')
axs[0].set_title('blue_channel')
axs[1].imshow(green_channel, interpolation='none')
axs[1].set_title('green_channel')
axs[2].imshow(red_channel, interpolation='none')
axs[2].set_title('red_channel')

# Adjust layout to prevent overlap
plt.tight_layout()
# Show the plot
plt.show()
```
