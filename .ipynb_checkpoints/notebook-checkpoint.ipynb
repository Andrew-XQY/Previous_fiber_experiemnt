# Configurations

```python
import pickle
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

'''Path to each folder and files'''

# path to calibrator
calibra = 'dataset/meas_14_7/data/DATA_MACHINE_7-14/MEAS_1/Cam_0_TS_1594729111252512_camTS6567617746.pk'
io_calibra_pairs = 'dataset/meas_14_7/calibration/calibration_15_7/-1.png'
```

# LED Screen

## Original image patterns
Original image patterns loaded on the Calibration LED screen?

```python
# Unpickling (deserializing)
with open(calibra, 'rb') as file:
    loaded_data = pickle.load(file)

# Convert the NumPy array to an image using PIL
image = Image.fromarray(loaded_data['img'])

# Display the image in the notebook using Matplotlib
plt.imshow(image)
plt.axis('off')  # To turn off the axis
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Replace 'your_image.png' with the path to your image file
image_path = io_calibra_pairs

# Open the image using PIL
image = Image.open(image_path)

# Display the image using Matplotlib
plt.imshow(image)
plt.axis('off')  # Turn off the axis
plt.show()

```
